Machine Learning

Proceso Estocastico:

- Un proceso estocastico P(x, y) genera pares de variables aleatorias x e y.

- Las caracteristicas x son facil de medir y suelen ser datos observables.

- Los targets (y) son mas dificiles de medir y pueden requerir metodos costosos o incluso ser imposible de obtener directamente (por ejemplo, la probabildiad de desarrollar una enfermedad terminal.)

Aprendizaje supervisado:

- El objetivo es construir un modelo matematico que prediga el valor de y en funcion de x.

- Esto se logra modelando la distribucion condicional P(y|x).

- Si se encuentra esta distribucion condicional, la estimacion de y basada en x es la esperanza condicional E[y/x].

- En los modelos parametricos, se asume una forma especifica para la distribucion condicional P(y|x, w). Donde w son los parametros del modelo.

- El objetivo es encontrar los parametros w que mejor describan la relacion entre x e y. Esto se logra maximizando la probabilidad conjunta de los datos observados bajo el modelo, es decir W* = argmaxw P(D|w), donde D = {(x1, y1), (x2, y2), .... , (xN, yN)} es el conjunto de datos.

- Probabilidad condicional P(y|x): Esta es la clave para hacer predicciones. Una vez que se tiene el modelo P(y|x, w), se puede calcular la probabilidad de diferentes valores de y dados nuevos valores de x.

- Estimacion de y: La prediccion final de y para un nuevo x a menudo se toma como el valor que maximiza P(y/x) (la moda de la distribucion condicional) o como la esperanza E[y|x].

Cuando trabajamos con datos tabulares, cada fila representa una observacion con sus caracteristicas (x) y su correspondiente target (y), es importante considerar las posibles fuentes de error y ruido que pueden afectar la calidad de nuestras predicciones y la precision del modelo.

1- Error de medicion:

- Sensado: El proceso de captura de datos puede introducir errores debido a la imprecision en los instrumentos de medicion o en el proceso de recoleccion de datos. Por ejemplo, un sensor de temperatura podria tener una precision limitada, lo que genera lecturas que se desvian de los valores reales.

- Error experimental: En un entorno experimental, puede haber errores inherentes al disenio o la ejecucion del experimento. Esto incluye errores humanos, problemas en la configutacion del experimento, o variaciones en las condiciones bajo las cuales se recolectaron los datos.

2- Ruido:

- Ruido aleatorio: El ruido aleatorio es una variacion impredecible en los datos que no sigue un patron especifico y puede ser causado por factores externos que afectan las mediciones de manera impredecible. Este ruido es inherentemente no sistematico y, en general, es dificil de eliminar completamente, pero tecnicas como el suavizado o la regularizacion pueden ayudar a mitigar su impacto.

- Ruido Sistematico: A diferencia del ruido aleatorio, el ruido sistematico es consitente y puede estar relaicoando con factores que no han sido correctamente controlados o medidos. Por ejemplo, si un sensor de presion siempre mide un valor ligeramente por debajo del real debido a un defecto en el equipo, ese sesgo se introducira de manera sistematica en los datos.

3- Variables omitidas:

- Factores no medidos: Puede haber variables importantes que no estan siendo medidas o incluidas en el modelo, y que tienen un impacto significativo en y. Por ejemplo, si estas prediciendo el rendimiento academico de estudiantes basado en horas de estudio y notas anteriores, pero no incluyes factores como el entorno familiar o el nivel de estres, tu modelo podria estar omitinedo informacion relevante.

- Variables latentes: Algunas variables pueden no ser directamente observables o medibles, pero influyen en la relacion entre x e y. Estas variables latentes pueden introducir complejidad adicional en el modelado, ya que sus efectos pueden ser confundidos con el ruido o con la influencia de otras variables observadas.

4- Problemas de colinealidad:

- Colinealidad entre variables: Si alguna de las caracteristicas x estan altamente correlacionadas entre si, puede ser dificil para el modelo distinguir sus efectos individuales sobre y. Esto puede introducir inestabilidad en las estimaciones de los coeficientes del modelo, especialmente en modelos lineales.

5- Errores de etiquetado:

- Errores en los targets: Es posible que los valores de y esten incorrectamente etiquetados debido a erorres humanos o a problemas en el proceso de recoleccion de datos. Estos errores pueden ser particularmente problematicos, ya que confunden al modelo durante el entrenamiento.

Mitigacion de errores:

- Preprocesamiento de datos: La limpieza y normalizacion de datos pueden ayudar a reducir el impacto de errores de medicion y ruido.

- Modelos robustos: Algunos algoritmos, como los arboles de decision y los metodos de ensamblado (e.g., random forest), son mas robustos frente a ruido en los datos.

- Regulizacion: Tecnicas como la regulizacion L1 o L2 pueden ayudar a reducir la sensibilidad del modelo a la colinealidad entre variables y a sobreajustes causados por el ruido.
