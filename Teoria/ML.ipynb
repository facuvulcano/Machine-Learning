ML MODELS


1- Linear Regression: Linear regression is a statistical method that models the relationship between a dependent variable (output) and one or mode independent variables (features) using a linear function. It is used to predict a continuos outcome.

- Mathematical Representation:

    y = β0 + β1x1 + β2x2 + … + βnxn + ϵ

    Where:

    y is the dependant variable (prediction/output)
    β0 is the intercept
    β1, β2, .... , βn are the coefficientes for each features x1, x2, .... , xn
    ϵ is the error term (residuals)

- Example:

Suppose we want to predict a person's weight based on their height. We might use a linear regression model where the weight is the dependent variable y and the height is the independent variable x. The model might look like:

weight = β0 + β1 x Height

2- Logistic Regression: Logisitc regression is a type of regression analysis used for binary classifcation problems. It estimates the probability that a given input point belongs to a certain class. Unlike linear regression, the ouput of logistic regresion is a probability value between 0 an 1.

- Mathematical representations:



- Example:

Suppose we want to predict wheather a customer will buy a product (yes or no) based on their age and income. Logistic regression can be used to estimate the probability that a customer will make a purchase. If the probability is greater than 0.5, we might classify the customer as likely to buy the product.

3- Perceptron: The perceptron is a type of artificial neural network and one of the simplest forms of a neural network model. It is mainly used for binary classification tasks. The perceptron model maps input features to a binary output.





4- Support Vector Machine (SVM): Is a powerful and versatile supervised learning model used for both classification and regression tasks. However, it is primarily used for binary classification. SVM aims to find the hyperplane that best separates the data into two classes.

Objective:

- Find the hyperplane that maximizes the margin, which is the distance between the hyperplane and the nearest data points from either class (called suport vectors)



5- k-Nearest Neighbors (KNN): K-Nearest Neighbors (KNN) is a simple, instance-based learning algorithm used for classification and regression. KNN workds by finding the k closest training examples in the feature space and making predictions based on the majority class (for classification) or the average value (for regression)

- Algorithm:

    1- Choose the number of neighbors k.
    2- Calculate the distance between the query example and all the training examples.
    3- Select the k nearest neighbors
    4- For classification, assign the class with the majority vote of the neighbors. For regression, compute the average of the k nearest neighbor's values.




6- Decision Trees: Is a non-parametric supervised learning algorithm used for both classification and regression tasks. It splits the data into subsests based on the value of input features, making decisions at each node until reaching a conclusion at the leaf nodes.

- Structure:

    - Root node: The topmost node that represents the entire dataset and feature.
    - Internal nodes: Nodes where the data is split based on a feature
    - Leaf nodes: Terminal nodes that represnt the output or final decision

7- Random Forest: Is an ensemble learning method that combines multiple decision trees to produce a more accurate and stable prediciton. It is used for both classification and regression tasks.

- Key concepts:

    - Boostrap Aggregating (Bagging): Each tree is trained on a different bootstrap sample of the data (a sample taken wiht replacement)
    - Feature randomness: At each split in the tree, a random subset of features is considered for splitting, introducing further randomness to help reduce overfitting.

- Prediciton:

    - Classification: Each tree votes for a class, and the majority vote is taken as the final prediction
    - Regression: The average of all tree predictions is taken as the final prediction

- Advantages:

    - Reduces overfitting by averaging multiple trees.
    - More robust and accuracte compared to a single decision tree.

8- Gradient Boosting: Is an ensemble learning technique that builds models sequenatially, with each model attempting to correct the errors of the previous ones. It is used for both classification and regression tasks.

- Key concepts:

    - Boosting: Combines weak learners (models that perform slightly bettern than random guessing) into a strong learner.
    - Gradient Descent: Each new model is trained to minimize the loss function, which measures the difference between the predicted and actual values, using gradient descent.

- Algorithm:

1- Initialize the model with a constant value, tyipically the mean of the target variable

2- Iterate over a number of boosting rounds:

    - Compute the pseudo-residuals, which are the negative gradients of the loss function with respect to the current model's predictions.

    - Train a warl learner (e.g., decision tree) on the pseudo-residuals

    - Update the model by adding the predictions of the weak learner scaled by a learning rate.

- Learning rate: A hyperparameter that control the contribution of each weak learner. Smaller values generally lead to better performance but requires more boosting rounds.

- Loss functions:

    - For regression: Mean squared error (MSE), Mean absolute error (MAE)
    - For classification: Logarithimic Loss (log loss), exponential loss

Overfitting and Underfitting

- Overfitting occurs when a model learns the detail and noise in the training data to the extent that it negatively impacts the performance of the model on the new data. This means the model is too complex, with too many parameteres realtive to the number of observations.

- Underfitting occurs when a model is too simple to learn the underlying pattern of the data. This often results from a model not having enough parameters or complexity to capture the underlying structure of the data.

Example of Overfitting

Imagine you're developing a machine learning model to predict house prices based on features like location, size, number of bedrooms, and age of the house. You decide to use a very complex model, such as a deep neural network with many layers and parameters.

Scenario:

- Training Data: You have a dataset of 100 houses with detailed features and prices.
- Model: A deep neural network with 10 layers.
- Result: The model performs exceptionally well on the training data, achieving nearly perfect predictions.

However, when you test this model on a new set of data (test data), the performance drops significally. The model has learned not only the genuine patterns from the training data but also the noise and specific quirks of the data set. This means it generalizes poorly to new, unseen data.

Example of Underfitting

Now, suppose you are tasked with the same prediction problem: forecasting house prices based on various features.

Scenario:

- Training Data: Same dataset of 100 houses.
- Model: You use a very simple linear regression model that only considers the size of the house (in square feet) to predict the price.
- Result: The model performs poorly even on the tranining data, failling to capture other critical factor influencing house prices, like location or number of bedrooms.

This model is too simple to capture the complexity and relationships in the data, resulting in high bias. It doesnt perform well on the training data and is unlinkely to do well on new data either.

How to address these issues

- Avoid overfitting: To prevent overfitting, you could simplify the model, use regularization techniques (like L1 or L2 regularization), or increase the amount of training data. Techniques lie cross-validations can also help ensure that the model generalizes well to new data.

- Avoid underfitting: To overcome underfitting, you might need to increase the model complexity. Consider using more features in the model or switching to a more sophisticated model that can captrue more complex relationships in the data. Adding more data can also help, provided the model is capable of learning deeper insights from that data.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import make_pipeline

# Generate some data
np.random.seed(0)
x = 2 - 3 * np.random.normal(0, 1, 20)
y = x - 2 * (x ** 2) + 0.5 * (x ** 3) + np.random.normal(-3, 3, 20)

# Reshaping the data
x = x[:, np.newaxis]
y = y[:, np.newaxis]

# Splitting the data into training and test sets
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0)

# Fitting a simple linear regression (underfitting)
model_underfit = LinearRegression()
model_underfit.fit(x_train, y_train)
y_train_pred_underfit = model_underfit.predict(x_train)
y_test_pred_underfit = model_underfit.predict(x_test)

# Fitting a high-degree polynomial regression (overfitting)
model_overfit = make_pipeline(PolynomialFeatures(degree=15), LinearRegression())
model_overfit.fit(x_train, y_train)
y_train_pred_overfit = model_overfit.predict(x_train)
y_test_pred_overfit = model_overfit.predict(x_test)

# Calculating the Mean Squared Error for both models
mse_train_underfit = mean_squared_error(y_train, y_train_pred_underfit)
mse_test_underfit = mean_squared_error(y_test, y_test_pred_underfit)
mse_train_overfit = mean_squared_error(y_train, y_train_pred_overfit)
mse_test_overfit = mean_squared_error(y_test, y_test_pred_overfit)

# Plotting the results
plt.figure(figsize=(14, 7))

# Plotting underfitting results
plt.subplot(1, 2, 1)
plt.scatter(x_train, y_train, color='red', label='Training data')
plt.scatter(x_test, y_test, color='blue', label='Test data')
plt.plot(x_train, y_train_pred_underfit, color='green', label='Model (Underfit)')
plt.title('Underfitting Scenario\nTrain MSE: {:.2f}, Test MSE: {:.2f}'.format(mse_train_underfit, mse_test_underfit))
plt.legend()

# Plotting overfitting results
plt.subplot(1, 2, 2)
plt.scatter(x_train, y_train, color='red', label='Training data')
plt.scatter(x_test, y_test, color='blue', label='Test data')
plt.plot(np.sort(x_train, axis=0), model_overfit.predict(np.sort(x_train, axis=0)), color='green', label='Model (Overfit)')
plt.title('Overfitting Scenario\nTrain MSE: {:.2f}, Test MSE: {:.2f}'.format(mse_train_overfit, mse_test_overfit))
plt.legend()

plt.show()

(mse_train_underfit, mse_test_underfit, mse_train_overfit, mse_test_overfit)
```

Metricas regresion:

Metricas:

1- True Positive (TP):

      - Definicion: Predicciones que son verdaderas y el modelo las clasifica correctamente como positivas.
      - Ejemplo: Si un modelo identifica correctamente a un paciente enfermo como enfermo

2- True Negative (TN):

      - Definicion: Predicciones que son falsas y el modelo las clasifica correctamente como negativas
      - Ejemplo: Si un modelo identifica correctamente a una persona sana como sana

3- False Negative (FN):

      - Definicion: Predicciones que son verdaderas pero el modelo las clasifica incorrectamente como negativas.
      - Ejemplo: Si un modelo identifica incorrectamente a un paciente enfermo como sano

4- Falso Positive (FP):

      - Definicion: Predicciones que son falsas pero el modelo las clasifica incorrectamente como positivas
      - Ejemplo: Si un modelo identifica incorrectamente a una pesona sana como enferma

5- Accuracy (Precision):

      - Definicion: Proporcion de predicciones correctas (tanto positicas como negativas) sobre el numero total de predicciones.
      - Formula: 
            Accuracy = (TP + TN) / (TP + TN + FP + FN)
      - Ejemplo: En un conjunto de datos con 8 predicciones, si 6 son correctas, la precision seria 6/8 = 75%

6- Precision (Precision Positiva):

      - Definicion: Proporcion de predicciones positivas correctas sobre el total de predicciones positivas realizadas por el modelo.
      - Formula:
            Precision = TP / (TP + FP)
      - Ejemplo: De todas las veces que el modelo predijo que alguien estaba enfermo, cuantas veces estuvo correcto

7- Recall (Sensibilidad o Tasa de Verdaderos Positivos)

      - Definicion: Proporcion de verdaderos positivos correctamente identificados sobre el total de positivos reales
      - Formula:
            Recall = TP / (TP + FN)
      - Ejemplo: De todas las personas enfermas en el conjunto de datos, cuantas fueron identificadas correctamente por el modelo.

8- False Positive Rate (Tasa de Falsos Positivos) (curva auc):
      
      - Definicion: Proporcion de negativos reales que son incorrectamente identificados como positivos
      - Formula:
            FPR = FP / (FP + TN)
      - Ejemplo: De todas las personas sanas en el conjunto de datos, cuantas fueron incorrectamente clasificads como enfermas por el modelo.

9- False Negative Rate (Tasa de Falsos Negativos)

      - Definicion: Proporcion de positivos reales que son incorrectmaente identificados como negativos
      - Formula:
            FNR = FN / (FN + TP)

K-Fold cross-validation: Ir validando de forma cruzada, en general k = 10

K-fold cross-validation is a popular method for assessing the performance of a machine learning model. It envolves splitting the dataset into k subsets (or "folds").

1- Split the data: Divide the datset into k equal (or nearly equal) parts. Each part is called a "fold".

2- Training and validation: For each of the k iterations:

- Use k - 1 folds for training the model.
- Use the remaining fold for validating the model.
- Rotate the validation fold in each iteration

3- Model evaluation: calculate the performance metric (e.g., accuracy, precision, recall, etc.) for each of the k iterations.

4- Average performance: Average the performance metrics from the k iterations to get a single estimate of the models performance.

Example: Supoose you have a dataet and choose k = 5 for 5-fold cross-validation:

1- Split the data: Divide the data into 5 folds (let's call them D1, D2, D3, D4 and D5).

2- Iterations:

- Iteration 1: Train on D2 + D3 + D4 + D5, validate on D1.

- Iteration 2: Train on D1 + D3 + D4 + D5, validate on D2.

- Iteration 3: Train on D1 + D2 + D4 + D5, validate on D3.

- Iteration 4: Train on D1 + D2 + D3 + D5, validate on D4.

- Iteration 5: Train on D1 + D2 + D3 + D4, validate on D5.

3- Performance metrics: Calculate the performance metric for each fold (e.g., accuracy).

4- Average performance: Average the five performance metrics to get an overall performance estimate.

Advantages:

- Reduces overfitting: By validating the model on different subsets, it helps ensure the model generalizes well to unseen data.

- Utilizes all data: Each data point is used both for training and validation, making better use of the available data.


```python
from sklearn.model_selection import cross_val_score
from sklearn.ensemble import RandomForestClassifier
from sklearn.datasets import load_iris

# Load a dataset
data = load_iris()
X, y = data.data, data.target

# Initialize the model
model = RandomForestClassifier()

# Perform k-fold cross-validation
k = 7
scores = cross_val_score(model, X, y, cv=k)

# Print the results
print(f"Cross-Validation Scores: {scores}")
print(f"Average Score: {scores.mean()}")
```

Dataset Construction

- Constructivsm in dataset construction involes creating datasets that accurately reflect the complexities and variabilities of the real world. This can include techniques like augmentation, ensuring diversity, and balancing classes to avoid biases.

Example of constructivism in dataset construction

- Scenario: Developing a facial recognition system for security pruposes.

- Objective: Create a facial recognition system that is robust, accurate, and can operate in various lightning conditions and diverse populations.

- Challenge: Facial recognition systems can often struggle with bias, particularly related to ethnicity, age, and lightning conditions, if not properly accounted for in the training data.

- Contructivist Approach:

1- Diversity in Ethnicity and age:
- Collect images from a wide variety of ethnic backgrounds and age groups to ensure the model learns to recognize faces across these diverse characteristics.
- Engage with communities worldwide to gather this data, ensuring consent and etichal use of data.

2- Varied Lightning and environmental conditions:
- Gather data in multiple lightining conditions ranging from bright sunlight to dimly lit environments to ensure the system is effective regardless of ligthing.
- Include images captured both indoors and outdoors, under various weather conditions.

3- Dynamic poses and expressions:
- Include images of individuals showing a range of facial expressions from different angles. This helps the model to recognize a person even when they are not facing directly towardds the camera or their facial expression changes.

4- Data agumentation:
- Use techniques like image rotation, scaling, cropping and lighting adjustments to artificially expand the dataset. This helps in training the model to be resilient to changes in perspective, size and ligthing.

5- Simulated data:
- Utilize CGI and advanced rendering techniques to simulate scenarios that are hard to capture in real life but are likelky to occur, thereby enriching the training dataset.

6- Feedback loops:
- Implement a system where the model's predictions are continuosly monitored, and feedback is used to further refine and expand the dataset. This could include scenarios where the model fails, which are then used to generate more data similar to the failure cases for retraining.

Benegits of the constructivist approach:

- Reduced Bias: By ensuring the dataset represents a broad range of human features, the system is less likely to exhibit biases toward any particular group.

- Enhanced robustness: Training with varied data improves the model's ability to function accurately in non-ideal conditions.

- Imporved generalization: The model is less likely to overfit to the specific characteristics of a narrowly defined training dataset and better at generalizing to new data similar yo real-world conditions.



Cross-validation

Is a powerful statistical technique used to evaluate the effectiveness of machine learning models. Its helps in understanding how well a model will generalize to an independent dataset. One of the main goals of using cross-validations is to mitigate problems like overfitting, ensuring the model's performance is reliable.

Types of cross-validation

1- K-Fold Cross-Validation

- The dataset is divided into 'k' consecutive folds.
- Each fold acts as the test set exactly once while the k-1 remaining folds form the training set.
- The process is repeated k times, with each of the k folds used exactly once as the test set.
- The average perfomance metric across all k trials is computed.

2- Stratified K-Fold Cross-Validation

- Similar to k-fold cross-validations but each fold contains roughly the same proportions of the target variable as the complete set.

- Especially useful for imbalanced datasets.

3- Leave-one-out Cros-Valudation (LOOCV)

- A sepcial case of k-fold cross-validation where k equals the number of data points in the dataset.
- For n data points, n separate times, a model is trained on all the data except for one point and tested on that single pint.
- Computationally expensive but can yield detailed insights, particulalry for samll datasets.

4- Group k-fold cross-validation

- Ensure that the same group is not represented in both testing and training sets.
- Useful for data where the samples are grouped (e.g., patients from several hospitals).

5- Time series Cross-Validation

- Data is split along the time axis, and the test sets are always a 'future' observation relative to the training set.
- Important for time series data where chronolocial order matters.
