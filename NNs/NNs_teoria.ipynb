En una red neuronal, una funcion de activacion tiene varios roles importantes, siendo el principal el de introducir no lineanlidad a la red. Esto permite a la red aprender y modelar relaciones complejas entre las variables de entrada y salida. Sin funciones de activacion no lineales, una red neuronal profunda se comportaria como una simple combinacion lineal de sus entradas, independientemente de la cantidad de capas ocultas que tenga.

Criterios para una funcion de activacion:

1- No linealidad: Una funcion de activacion debe ser no lineal para que la red pueda aprender patrones complejos. Esto permite a la red capturar las relaciones no lineales en los datos.

2- Derivabilidad: La funcion de activacion debe ser diferenciable en casi todos sus puntos. Esto es crucial para el proceso de backpropagation, que se utiliza para ajustar los pesos de la red durante el entrenamiento.

3- Rango de salida: El rango de salida de la funcion de activacion puede influir en el comportamiento de la red. por ej:
    
    - Sigmoide: rango entre (0, 1), util para clasif binaria
    - tanh: rango entre (-1, 1), util para centrarse en datos normalizados.
    - Relu: rango [0, +inf)

4- Eficiencia computacional: La funcion de activacion debe ser computacionalmente eficiente para que el entrenamiento y la inferencia de la red sean rapidos. Funciones como ReLU son populares debido a su simiplicidad y velocidad de calculo.

5- Evitar saturacion: Algunas funciones de activacion, como la sigmoide y tanh, pueden saturarse en sus extremos, lo que lleva a gradientes muy pequenios y puede relentizar ele entranmiento.
